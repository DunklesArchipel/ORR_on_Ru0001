# Unitpackage usage

The [unitpackage](https://echemdb.github.io/unitpackage/) can be used to explore the data in the data folder for further data evaluation or retrieving system specific information.


```python
from unitpackage.cv.cv_collection import CVCollection
from unitpackage.local import collect_datapackages

db = CVCollection(collect_datapackages('../data/'))
db['figure_2_orr_hclo4_b'].plot()
```

## Query entry specific information

```python
db['figure_2_orr_hclo4_b'].system.electrolyte.components
```

```python
db['figure_2_orr_hclo4_b'].system.electrodes
```

```python
db['figure_2_orr_hclo4_b'].get_electrode('WE').preparation_procedure.description
```

## rescale the data

```python
db['figure_2_orr_hclo4_b'].rescale({'j':'mA cm-2'}).plot()
```
